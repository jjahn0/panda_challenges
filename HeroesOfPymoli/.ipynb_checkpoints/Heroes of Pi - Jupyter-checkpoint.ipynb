```python
player_file = os.path.join('generated_data','players_complete.csv')
players_df = pd.read_csv(player_file, encoding='ISO-8859-1')
players_df.head()
```

```python
# total number of players
player_count = players_df['Player ID'].count()
player_count
```

```python
item_file = os.path.join('generated_data', 'items_complete.csv')
items_df = pd.read_csv(item_file, encoding='ISO-8859-1')
items_df.head()
```

```python
#unique item count
item_count = items_df['Item ID'].count()
item_count
```

```python
purchase_file = os.path.join('generated_data', 'purchase_data_3.csv')
purchase_df = pd.read_csv(purchase_file, encoding='ISO-8859-1')
purchase_df.head()
```

```python
# total number of purchases
purchase_num = purchase_df['Purchase ID'].count()
purchase_num
```

```python
# average purchase price
purchase_avg = purchase_df['Price'].mean()
purchase_avg
```

```python
# total revenue
total_revenue = purchase_df['Price'].sum()
total_revenue
```

```python
purchase_data = {'Unique Items': [item_count],
        'Total Purchases': [purchase_num],
        'Avg. Purchase price': [purchase_avg],
        'Total Revenue': [total_revenue]
       }

purchase_table = pd.DataFrame(data=purchase_data)
purchase_table
```

```python
gender_count = pd.DataFrame(players_df['Gender'].value_counts())
gender_count
```

```python
gender_percent = (gender_count['Gender'] / gender_count['Gender'].sum())*100
gender_count['Gender Percent'] = gender_percent
gender_count['Gender Percent'] = gender_count['Gender Percent'].map('{:.2f}%'.format)
gender_count
```

```python
group_gender = purchase_df.groupby(['Gender'])
purchase_group_gender = group_gender['Price'].sum()
gender_dem_df = pd.DataFrame(purchase_group_gender)
gender_dem_df = gender_dem_df.rename( columns={'Price':'Revenue'})
gender_dem_df
```

```python
purchase_total = group_gender['Purchase ID'].count()
purchase_total
```

```python
gender_dem_df['Purchase Count'] = purchase_total
gender_dem_df
```

```python
gender_dem_df['avg. purchase price'] = gender_dem_df['Revenue']/gender_dem_df['Purchase Count']
gender_dem_df['avg. purchase price'] = gender_dem_df['avg. purchase price'].map('${:.2f}'.format)
gender_dem_df['Revenue'] = gender_dem_df['Revenue'].map('${:.2f}'.format)
gender_dem_df
```

```python
bins = [0,10,15,25,35,45]
group_names=['child','youth','young adult','adult','middle-aged']
purchase_df['Age Category'] = pd.cut(purchase_df['Age'], bins, labels=group_names)
age_grouped_df = purchase_df.groupby('Age Category')
age_grouped_df.head()
```

```python
age_cat_df = pd.DataFrame(age_grouped_df['Purchase ID'].count())
age_cat_df = age_cat_df.rename(columns = {'Purchase ID': 'Purchase Count'})
age_cat_df
```

```python
age_revenue = age_grouped_df['Price'].sum()
age_cat_df['Revenue'] = age_revenue
age_cat_df
```

```python
avg_age_revenue = age_cat_df['Revenue'] / age_cat_df['Purchase Count']
age_cat_df['avg. purchase price'] = avg_age_revenue
age_cat_df['avg. purchase price'] = age_cat_df['avg. purchase price'].map('${:.2f}'.format)
age_cat_df['Revenue'] = age_cat_df['Revenue'].map('${:.2f}'.format)
age_cat_df
```

```python
sn_price_df = purchase_df[['SN', 'Price']]
top_df = sn_price_df.groupby('SN')
top_whales_df = pd.DataFrame(top_df['SN'].count())
top_revenue = top_df['Price'].sum()
top_whales_df = top_whales_df.rename(columns={'SN':'Purchase Count'})
top_whales_df['Revenue'] = top_revenue
top_avg_purchase = top_whales_df['Revenue']/top_whales_df['Purchase Count']
top_whales_df['Avg. Purchase Price'] = top_avg_purchase
top_5_df = top_whales_df.sort_values('Revenue', ascending=False).head()
top_5_df['Revenue'] = top_5_df['Revenue'].map('${:.2f}'.format)
top_5_df['Avg. Purchase Price'] =  top_5_df['Avg. Purchase Price'].map('${:.2f}'.format)
top_5_df
```
